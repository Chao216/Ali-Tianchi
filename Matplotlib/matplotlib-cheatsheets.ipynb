## 1 Quick start

```python
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 2*np.pi, 100)
Y = np.cos(X)
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y, color="g")
```

```python
fig.savefig("./Figures/cosx.pdf")
fig.show()
```

## 2 subplot

### 2.1 object-oriented api

```python
fig, axs = plt.subplots(2,3) # 2 rows, 3 columns
axs[0,0].plot(X,Y)
axs[0,1].plot(X,Y, color="r")
axs[0,2].plot(X,Y, color="orange")
axs[1,0].plot(X,Y, color="y")
axs[1,1].plot(X,Y, color="g")
axs[1,2].plot(X,Y, color="purple")

fig.suptitle("index based oop")

fig.savefig("./Figures/colored_subs.pdf")
```

```python
fig, ((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(2,3)
ax1.plot(X,Y, ls="-")
ax2.plot(X,Y, ls="--")
ax3.plot(X,Y, ls=":")
ax4.plot(X,Y, ls="-.")
ax5.plot(X,Y, ls="-.", color="g")
ax6.plot(X,Y, ls="--", color="r")

fig.suptitle("destructed oop")

fig.savefig("./Figures/linestyle_subs.pdf")
```

### 2.2 matlab style api

```python
plt.figure()

plt.subplot(2,3,1)
plt.plot(X,Y)

plt.subplot(2,3,2)
plt.plot(X,Y, color="r")

plt.subplot(2,3,3)
plt.plot(X,Y, color="orange")

plt.subplot(2,3,4)
plt.plot(X,Y, color="y")

plt.subplot(2,3,5)
plt.plot(X,Y, color="g")

plt.subplot(2,3,6)
plt.plot(X,Y, color="purple")

plt.suptitle("Matlab style ploting")

plt.savefig("./Figures/Matlab-style.pdf")
```

## 3 basic plots

```python
X = np.linspace(1,100,100)
fig, ax = plt.subplots(5,2, figsize=(10,10))

ax[0,0].plot(X,X, label="Linear")
ax[0,0].legend()

ax[0,1].scatter(np.random.randint(1,100,100),np.random.randint(1,100,100), label="scatter")
ax[0,1].legend()

ax[1,0].bar(X, np.random.randint(1,100,100), color="orange", label="bar-plot")
ax[1,0].legend()

sizes = [10,20,50,120,30]
explodes = [0.1,0.2,0,0.1,0]
labels = ["amy","Tony","larry","eva","adam"]

ax[1,1].pie(sizes, explode=explodes, labels= labels)
ax[1,1].axis("equal")

x = np.linspace(0, 2*np.pi,100)

ax[2,0].fill_between(x, np.sin(x)+1,np.sin(x)-1)
ax[2,0].set_title("fill")

ax[2,1].text(0.5,0.5,"Hello, this is a text", rotation = 45)


x_features = np.linspace(-5,5,100)
y_features = np.linspace(-5,5,100)
X,Y = np.meshgrid(x_features, y_features)

Z = X**2+Y**2

ax[3,0].contour(X,Y,Z)
ax[3,0].set_title("contour plot")
ax[3,0].axis("equal")

ax[3,1].pcolormesh(X,Y,Z)
ax[3,1].set_title("pcolormesh")
ax[3,1].axis("equal")

ax[4,0].quiver(0,0,1,1)
ax[4,0].set_title("quiver vector")

grid = np.random.rand(10,10)
ax[4,1].imshow(grid)
ax[4,1].set_title("imshow")

fig.subplots_adjust(wspace=0.2,hspace=0.7)

fig.suptitle("basic plots")

fig.savefig("./Figures/basicPlots.pdf")
```
