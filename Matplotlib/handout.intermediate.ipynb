```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(3,3)
fig.subplots_adjust(wspace =0.2, hspace=0.5)
ax[0,0].set_facecolor("#ddddff")
ax[2,2].set_facecolor("#ffffdd")
```

```python
fig = plt.figure()

gs = fig.add_gridspec(4,4)

ax1 = fig.add_subplot(gs[0:2,:])
ax1.set_facecolor("red")
ax2 = fig.add_subplot(gs[2:4,0:1])
ax2.set_facecolor("orange")
ax3 = fig.add_subplot(gs[2:3,1:3])
ax3.set_facecolor("yellow")
ax4 = fig.add_subplot(gs[3:4,3:4])
ax4.set_facecolor("green")
ax4.set_title("you deserve a title")

fig.suptitle("divide with gridspec")
```

```python
fig, ax = plt.subplots(2,2)

ax[0,0].spines["top"].set_color("None")
ax[0,1].spines["bottom"].set_color("None")
ax[1,0].spines["left"].set_color("None")
ax[1,1].spines["right"].set_color("None")
```

```python
from matplotlib.ticker import MultipleLocator as ML
from matplotlib.ticker import ScalarFormatter as SF

# change x y lim and scale 
fig, ax = plt.subplots()

ax.set_xlim(0,100)
ax.xaxis.set_minor_locator(ML(5))
ax.xaxis.set_minor_formatter(SF())

ax.set_ylim(0,1000)
ax.yaxis.set_minor_locator(ML(40))
ax.yaxis.set_minor_formatter(SF())

# rotate ticks
ax.tick_params(axis="x", which="minor", rotation=90)
ax.tick_params(axis="x", which="major", rotation=-15)
```

```python
fig, ax = plt.subplots()
x=np.linspace(0,2*np.pi,100)
y = np.sin(x)
ax.plot(x,y)
ax.set_xlim(0,2*np.pi)
ax.set_ylim(-1.3,1.3)
ax.annotate('local max', xy=(0.5*np.pi, 1),  xycoords='data',
            xytext=(3, 1.2), textcoords='data',
            arrowprops=dict(facecolor='pink', alpha = 0.3,width=3, headwidth=5, headlength=6 ),
            horizontalalignment='right', verticalalignment='top',
            )
ax.annotate('local min', xy=(1.5*np.pi, -1),  xycoords='data',
            xytext=(5, -0.3), textcoords='data',
            arrowprops=dict(facecolor='pink', alpha = 0.3,width=3, headwidth=5, headlength=6 ),
            horizontalalignment='right', verticalalignment='top',
            )

fig.savefig("./Figures/annotation.pdf")
```
