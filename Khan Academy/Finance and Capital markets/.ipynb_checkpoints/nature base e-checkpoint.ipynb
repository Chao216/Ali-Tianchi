```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,100000001,1)
```

```python
x
```

```python
y = list(map(lambda n:(1+1/n)**n,x))
```

```python
y
```

```python
plt.plot(x,y)
```

```python

fig, ax = plt.subplots()
ax.plot(x[:100],y[:100],label="approximation curve")
ax.set_title("(1+1/n)^n ")
ax.set_xlabel("n")
ax.set_ylabel("value")
ax.legend()
fig.savefig("./Images/nature_base_e.pdf")
```
