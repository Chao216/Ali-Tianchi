```python
x = np.arange(1,8360,1)
```

```python
df = pd.DataFrame({"Date":dates,"value":y})
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df.Date,df.value)
```

```python
z = list(map(lambda x: x*0.7,x))
```

```python
z
```

```python
df["values"]= z
```

```python
df
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(df.Date,df.value)
ax[1].plot(df.Date,df.values)

fig.savefig("./data/1_2_subplots.pdf")
```
