# 创建 MultiIndex

```python
L1 = ["A","A","A","A","B","B","B","B","C","C","C","C",]
L2 = ["a","b","c","d"]*3

MI1 = pd.MultiIndex.from_arrays([L1,L2],names=["UPPER","lower"])
MI1
```

```python
tuples = [("red",1),("red",2),("red",3),("red",4),("red",5),("red",6),("red",7),("orange",1),("orange",2),("orange",3),("orange",4),("orange",5),("orange",6),("orange",7)]

MI2 = pd.MultiIndex.from_tuples(tuples, names=["color","number"])
MI2
```

# 索引层 交换与删除

```python
np.random.seed(0)
L1,L2,L3 = ['A','B'],['a','b'],['alpha','beta']
mul_index1 = pd.MultiIndex.from_product([L1,L2,L3], names=('Upper', 'Lower','Extra'))
L4,L5,L6 = ['C','D'],['c','d'],['cat','dog']
mul_index2 = pd.MultiIndex.from_product([L4,L5,L6], names=('Big', 'Small', 'Other'))
df_ex = pd.DataFrame(np.random.randint(-9,10,(8,8)), index=mul_index1,  columns=mul_index2)
df_ex
```

## swap level

```python
df_ex.swaplevel(0,1,axis=0)
```

```python
df_ex.swaplevel(0,2,axis=1)
```

## reorder levels

```python
df_ex.reorder_levels([2,1,0],axis=1)
```

```python
df_ex.reorder_levels([2,1,0], axis=0)
```

# 删除一层，多层索引

```python
df_ex.droplevel(2,axis=0)
```

```python
df_ex.droplevel([0,1],axis=1)
```

## 修改索引层 名字

```python
df_ex.rename_axis(index={"Upper":"大写","Lower":"小写"}, columns={"Big":"大的","Small":"小的"})
```

## rename 修改索引值

```python
df_ex.rename(index={"A":"啊","B":"波"}, columns={"C":"慈","D":"得"}, level=0)
```

# 索引的运算



```python
df_set_1 = pd.DataFrame([[0,1],[1,2],[3,4]], index = pd.Index(['a','b','a'],name='id1'))
df_set_2 = pd.DataFrame([[4,5],[2,6],[7,1]], index = pd.Index(['b','b','c'],name='id2'))
```

```python
df_set_1
```

```python
df_set_2
```

```python
id1, id2 = df_set_1.index.unique(), df_set_2.index.unique()
```

```python
id1
```

```python
id2
```

## intersection union difference symmetric_difference

```python
id1.intersection(id2)
```

```python
id1.union(id2)
```

```python
id1.difference(id2)
```

```python
id1.symmetric_difference(id2)
```
