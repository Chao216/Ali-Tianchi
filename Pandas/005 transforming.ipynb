# pivot 

# Long ----> Wide

```python
df = pd.DataFrame({'Class':[1,1,2,2],
                   'Name':['San Zhang','San Zhang','Si Li','Si Li'],
                   'Subject':['Chinese','Math','Chinese','Math'],
                   'Grade':[80,75,90,85]})
df
```

```python
df.pivot(index="Name",columns="Subject",values="Grade")
```

```python
df.pivot(index="Class",columns="Subject",values="Grade")
```

```python
df = pd.DataFrame({'Class':[1, 1, 2, 2, 1, 1, 2, 2],
                   'Name':['San Zhang', 'San Zhang', 'Si Li', 'Si Li',
                              'San Zhang', 'San Zhang', 'Si Li', 'Si Li'],
                   'Examination': ['Mid', 'Final', 'Mid', 'Final',
                                    'Mid', 'Final', 'Mid', 'Final'],
                   'Subject':['Chinese', 'Chinese', 'Chinese', 'Chinese',
                                 'Math', 'Math', 'Math', 'Math'],
                   'Grade':[80, 75, 85, 65, 90, 85, 92, 88],
                   'rank':[10, 15, 21, 15, 20, 7, 6, 2]})
df
```

```python
df.pivot(index=["Name","Examination"],columns="Subject",values=["Grade","rank"])
```

# pivot table

```python
df = pd.DataFrame({'Name':['San Zhang', 'San Zhang', 
                              'San Zhang', 'San Zhang',
                              'Si Li', 'Si Li', 'Si Li', 'Si Li'],
                   'Subject':['Chinese', 'Chinese', 'Math', 'Math',
                                 'Chinese', 'Chinese', 'Math', 'Math'],
                   'Grade':[80, 90, 100, 90, 70, 80, 85, 95]})
df
```

## aggfunc 表示对重复 index 的column 值进行 聚合操作，例如 max, min, mean

```python
df.pivot_table(index="Name",columns="Subject",values="Grade",aggfunc="mean")
```

```python
df.pivot_table(index="Name",columns="Subject",values="Grade",aggfunc="max")
```

```python
df.pivot_table(index = 'Name',
               columns = 'Subject',
               values = 'Grade',
               aggfunc = lambda x:x.mean())
```

## pivot_table margins 表示汇总，列和行

```python
df.pivot_table(index = 'Name',
               columns = 'Subject',
               values = 'Grade',
               aggfunc='mean',
               margins=True)
```

# melt

# wide ----> long

```python
df = pd.DataFrame({'Class':[1,2],
                   'Name':['San Zhang', 'Si Li'],
                   'Chinese':[80, 90],
                   'Math':[80, 75]})
df
```

```python
df.melt(id_vars=["Class","Name"],value_vars=["Chinese","Math"],var_name="Subject",value_name="Grade")
```

### id_vars 原始 做id功能的变量， value_vars 要变成值的变量，新column的列名（变量名）， 原变量的值的新变量名。

# ---------------------------------------------------------------------------------------------------------------------------------------

```python
long = pd.DataFrame({'Class':[1,1,2,2],
                   'Name':['San Zhang','San Zhang','Si Li','Si Li'],
                   'Subject':['Chinese','Math','Chinese','Math'],
                   'Grade':[80,75,90,85]})
long
```

```python
wide = pd.DataFrame({'Class':[1,2],
                   'Name':['San Zhang', 'Si Li'],
                   'Chinese':[80, 90],
                   'Math':[80, 75]})
wide
```

```python
long.pivot(index=["Class","Name"],columns="Subject",values="Grade")
```

```python
wide.melt(id_vars=["Class","Name"],value_vars=["Chinese","Math"],var_name="科目",value_name="成绩")
```

# wide to long

```python
df = pd.DataFrame({'Class':[1,2],'Name':['San Zhang', 'Si Li'],
                   'Chinese_Mid':[80, 75], 'Math_Mid':[90, 85],
                   'Chinese_Final':[80, 75], 'Math_Final':[90, 85]})
df
```

```python
pd.wide_to_long(df=df, stubnames=["Chinese","Math"],i=["Class","Name"],j="Exam",sep="_",suffix=".+")
```

```python
pd.wide_to_long(df=df, stubnames=["Chinese","Math"],i=["Class","Name"],j="考试",sep="_",suffix=r"[a-zA-Z0-9]*")
```

## df 要变换的表，stubnames 要处理列名前一部分，i index列，j 新列（index）的名称，sep 分隔符，suffix 处理列名的后缀部分 【可以用正则表达式】。

# 索引 变形  stack 与 unstack

```python
df = pd.DataFrame(np.ones((4,2)),
                  index = pd.Index([('A', 'cat', 'big'),
                                    ('A', 'dog', 'small'),
                                    ('B', 'cat', 'big'),
                                    ('B', 'dog', 'small')]),
                  columns=['col_1', 'col_2'])
df
```

## 把行索引 变成 列索引

```python
df.unstack()
```

```python
df.unstack(0)
```

```python
df.unstack([0,2])
```

# 把列索引变成行索引

```python
df = pd.DataFrame(np.ones((4,2)),
                  index = pd.Index([('A', 'cat', 'big'),
                                    ('A', 'dog', 'small'),
                                    ('B', 'cat', 'big'),
                                    ('B', 'dog', 'small')]),
                  columns=['index_1', 'index_2']).T
df
```

```python
df.stack()
```

```python
df.stack(0)
```

```python
df.stack([0,2])
```
