```python
import numpy as np
import pandas as pd
```

# df.groupby(分组依据)[数据来源].使用操作

```python
df = pd.read_csv("./data/learn_pandas.csv")
```

```python
df.head()
```

## 各校平均身高

```python
df.groupby("School")["Height"].mean()
```

## 各校 男女的平均体重

```python
df.groupby(["School","Gender"])["Weight"].mean()
```

## condition for groupby

```python
condition = df.Weight >= df.Weight.median()
```

```python
condition
```

```python
df.groupby(condition)["Height"].median()
```

### 体重不超过中位数的一组，身高中位数157， 体重超过中位数的另一组，身高中位数167.

# Groupby 对象

```python
gb = df.groupby(["School","Grade"])
```

```python
gb.ngroups
```

```python
results = gb.groups
results.keys()
```

```python
gb.size()
```

```python
gb.get_group(("Fudan University","Senior"))
```

# 聚合函数 Aggregation

```python
gb= df.groupby("Gender")["Weight"]
```

```python
gb.mean()
```

```python
gb.idxmax()
```

```python
gb.idxmin()
```

# quantile 百分位

```python
gb.quantile([0.25,0.5,0.75])
```

# 分组后，接受多个来源数据

```python
gb = df.groupby(["School","Gender"])[["Height","Weight"]]
```

```python
gb.max()
```

# agg 方法

##  使用多个函数

```python
gb.agg(["min","max","median"])
```

## 对特定的列使用特定的聚合函数

```python
gb.agg({"Height":["min","median","max"],"Weight":["mean","idxmin","idxmax"]})
```

## 使用自定义函数

```python
gb.agg(lambda x : x.max()-x.min())
```

## 聚合结果重命名

```python
gb.agg([("Peak to Peak", lambda x: x.max()-x.min()),("Standard Deviation", "std"),("Vardiance","var")])
```

# 变换函数 与transform 方法。

```python
gb.cummax()
```

### 把每位同学的身高体重替换成所在组的最小值。

```python
gb.transform(lambda x: x.min()) 
```

### 把每位同学的身高体重替换成所在组的平均值。

```python
gb.transform(lambda x: x.mean()) 
```

# 组索引过滤

```python
gb.filter(lambda x: x.shape[0]>45)
```

```python
gb.filter(lambda x : x.shape[0]>47)
```

# 以上保留下来的事分组后，组员属大于47的。

```python
gb.size()
```

# 跨列分组 apply

```python
def BMI(x):
    Height = x['Height']/100
    Weight = x['Weight']
    BMI_value = Weight/Height**2
    return BMI_value.mean()
gb.apply(BMI)
```

```python
gb = df.groupby("School")[["Height","Weight"]]
```

# 返回标量

```python
gb.apply(lambda x: x.mean())
```

```python
gb.apply(lambda x: np.nan)
```

# 返回 Series

```python
gb.apply(lambda x : pd.Series(data=x.mean(),index=["Height","Weight"],name="A series"))
```

```python
gb.apply(lambda x : pd.Series(data=x.mean(),name="A series"))
```

# 返回 dataframe

```python
gb.apply(lambda x: pd.DataFrame(data=x.min(), columns=["Min Value"]))
```
