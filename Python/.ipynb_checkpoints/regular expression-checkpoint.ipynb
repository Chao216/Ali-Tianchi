```python
import re
```

```python
sherlock = open("./Files/pg2097.txt", encoding="UTF-8").read()
```

```python
print(sherlock)
```

```python
pattern = re.compile(r"CRIMINAL",re.I)
```

```python
mo = pattern.findall(sherlock)
```

```python
for i in mo :
    print(i)
```

```python
patn2 = re.compile(r"bad", re.I)
mo2 = patn2.findall(sherlock)
```

```python
for t in mo2:
    print(t)
```
