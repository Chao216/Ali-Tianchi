### 假设每小时卖出去13瓶可乐，那么十分钟内会卖出去多少可乐？

```python
lam = 13 / 6
```

```python
lam
```

lambda 表示单位时间内发生时间的次数

```python
x = np.random.poisson(lam, 10000)
```

```python
plt.hist(x,bins =20)
```

## 20至动物里7只是小狗，随便抽取12只动物。抽到不同数量小狗的概率分布

```python
x = np.random.hypergeometric(7,13,12,10000)
```

```python
plt.hist(x,bins=12)
```
